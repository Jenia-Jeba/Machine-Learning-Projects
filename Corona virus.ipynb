```python
# made list of dupplicate and unique_columns

duplicates_columns = []
unique_columns =[]
for item in columns_name:
    if columns_name.count(item) > 1:
        duplicates_columns.append(item)
    else:
        unique_columns.append(item)
        
duplicates_columns = list(set(duplicates_columns))

columns_name = duplicates_columns + unique_columns
```

```python
#df columns

def build_columns_df(datasets,columns_df):
    """
     made df dataset VS new dataset column
    """
    columns_name_value = []

    for j in columns_name:
        x =[]
        for i in datasets:
            x.append(i.keys().contains(j))
        columns_name_value.append(x)

    dict = {i:x for i,x in zip(columns_name,columns_name_value)}
    columns_df = pd.DataFrame(dict)

    columns_df.rename(index= {i:x for i,x in zip(np.arange(0,len(datasets_name)),datasets_name)}, inplace = True)
    return columns_df
```

```python
#unique_columns by df

def unique_columns_df(unique_columns, df):
    for j in unique_columns:
        if df.keys().contains(j) == True:
            print(j)
```
